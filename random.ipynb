```python
import  pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
```

# IMPORTING THE DATASET :

```python
data = pd.read_csv('campus_placement.csv')
data.drop(['salary','hsc_s'], axis=1, inplace=True)
```

```python
data.head()
```

# CLEANING THE DATA :

```python
male = pd.get_dummies(data['gender'], drop_first=True)
ssc_board = pd.get_dummies(data['ssc_b'], drop_first=True)
hsc_board = pd.get_dummies(data['hsc_b'], drop_first=True)
degree = pd.get_dummies(data['degree_t'], drop_first=True)
work_exp = pd.get_dummies(data['workex'], drop_first=True)
specs = pd.get_dummies(data['specialisation'], drop_first=True)
hired_or_not = pd.get_dummies(data['status'], drop_first=True)
```

```python
hired_or_not.head()
```

```python
data = pd.concat([data,male,ssc_board,hsc_board,degree,work_exp,specs,hired_or_not], axis=1)
```

```python
data.head()
```

```python
data.head()
```

```python
data.drop(['gender','ssc_b','hsc_b','degree_t','workex','specialisation','status'], axis=1, inplace=True)
```

```python
data.head()
```

# TRAINING THE DATA :
### HERE THE TRAINING SIZE IS GIVEN TO BE 30%

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('Placed', axis=1)
y = data['Placed']

x.shape
y.shape

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

# MODEL SELECTION :
### ( LOGISTIC REGRESSION HAS BEEN USED )

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

# PREDICTING THE VALUES :

```python
pred = log_reg.predict(x_test)
```

# CHECKING THE ACCURACY OF THE MODEL :

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```
